```python
'''
What we will do
- Tokenize each word
- POS_tag the tokens
- Use NER - Named entity recognition, i.e., all the proper names of skills, person, organization will be stored,
  then we can simply add these elements to some set. The reason for taking set is that set does not allow duplicate
  values, hence we can be sure that element is not stored more than once.
  Then we will convert that set to list and list to a dataframe
'''



x = set()

for i in rsd['RAW DATA']:
    ans = []
    s = nltk.pos_tag(word_tokenize(i))
    ne_ch = nltk.ne_chunk(s)
    
    for tagged_tree in ne_ch:
        #here see and apply NER, store named entities
        if hasattr(tagged_tree, 'label'):
            entity_name = ' '.join(c[0] for c in tagged_tree.leaves())
            entity_type = tagged_tree.label()
            x.add(entity_name)
            #print(entity_name,"   ",entity_type,"\n")
            
#set to list
lis = list(x)
#list to dataframe
df = pd.DataFrame(lis)

df
```
